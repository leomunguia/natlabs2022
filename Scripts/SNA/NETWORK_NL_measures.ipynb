```python
#Conexión general de la red
nx.degree(red_ln)
```

```python
#conexiones locales de la red
nx.degree_centrality(red_ln)
sorted(nx.degree_centrality(red_ln).values())
```

```python
ln_central = nx.degree_centrality(red_ln)
```

```python
for i in sorted(ln_central, key=ln_central.get, reverse=True):
    print(i, ln_central[i])
    
    
```

```python
plt.figure(figsize=(20,15))
plt.hist(nx.centrality.degree_centrality(red_ln).values(), color="green");
plt.title("Degree centrality of laboratories network \n", size=25)
plt.grid()
plt.xticks(fontsize = 25) 
plt.yticks(fontsize = 22)
plt.xlabel("\n Centrality degree", size=25)
plt.ylabel("Nodes", size=25)
plt.savefig("Degree centrality of laboratories network.svg")
```

```python
relacion_importante = nx.eigenvector_centrality(red_ln)
```

```python
for i in sorted(relacion_importante, key=relacion_importante.get, reverse=True):
    print(i, relacion_importante[i])
```
